```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.linear_model import LinearRegression

df = pd.read_csv('es.csv')
df['Release Date']= pd.to_datetime(df['Release Date'],format='%Y')

df
```

```python
def input_format(data): return data.map(datetime.toordinal).values.reshape(-1, 1)

m = LinearRegression()
m.fit(input_format(df['Release Date']),df['Number']);
```

```python
prediction_range = pd.date_range('1994', '2025', freq='AS')

plt.plot(prediction_range,m.predict(input_format(prediction_range)),df['Release Date'],df['Number'],'ro')
plt.title('How many Elder Scrolls games are we currently owed?');
```
